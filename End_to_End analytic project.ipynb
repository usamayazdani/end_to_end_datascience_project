```python
#AS there is not available,unknown value in dataset we make it nan 
df=pd.read_csv('orders.csv', na_values=['Not Available','unknown'])
df['Ship Mode'].unique()
```

```python
#rename column to make it same order in lower_Case
df.columns=df.columns.str.lower() #get into list string and apply list function on it
df.columns=df.columns.str.replace(' ','_')
```

```python
#create new column 
df['discount']=df['list_price']*df['discount_percent']*0.1 #to get the dicount
df['sale_price']=df['list_price']-df['discount']  #how much 
df['profit']=df['sale_price']-df['cost_price']  #how much profit we earn from it
df
```

```python
df.dtypes
```

```python
#convert data column into date format
df['order_date']=pd.to_datetime(df['order_date'],format="%Y-%m-%d")
```
